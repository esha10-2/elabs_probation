```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r"C:\Users\KIIT\Downloads\Sample - Superstore(2).csv"
try:
    df= pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    df= pd.read_csv(file_path, encoding='latin1')
subcat_sales=df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
plt.bar(subcat_sales.index, subcat_sales.values, width=0.5, color='brown')
plt.xticks(rotation=45)
plt.xticks(fontsize=8)
plt.xlabel('Sub Category')
plt.ylabel('Sales')
plt.title('Bar Graph For Sub-Category vs Sales')
plt.show()
```

The bar graph above shows the sales of each sub-category of products. From the graph, we get the following insights: 
1. Phones and chairs have the largest numbers of sales, indicating that customers spend the most in these two sub-categories.
2. Storage, tables and binders follow, bringing in decent amounts of sales and contributing moderately to the total sales.
3. Items like envelopes, labels and fasteners are at the bottom of the list indicating that they generate low demand.

Overall, the graph emphasizes that the company depends largely on the sales of phones and chairs while items like labels and fasteners contribute little to the revenue.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r"C:\Users\KIIT\Downloads\Sample - Superstore(2).csv"
try:
    df= pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    df= pd.read_csv(file_path, encoding='latin1')
category_counts = df['Category'].value_counts()
plt.pie(category_counts, labels=category_counts.index, startangle=90, colors=['pink','brown','purple'])
plt.title('Pie Chart for Categories')
plt.legend(loc='lower right')
plt.show()
```

1. The office supplies which consist of items like binders, paper, storage, art, envelopes, fastners, etc. take up the largest portion of shares meaning that this category generates the largest revenue.
2. The other two categories contribute almost equally to the shares.

The pie chart shows that office supplies dominate in volume and generate the largest revenue while furniture and technology occupy almost equal volumes and contribute almost equally to the revenue.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r"C:\Users\KIIT\Downloads\Sample - Superstore(2).csv"
try:
    df= pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    df= pd.read_csv(file_path, encoding='latin1')
sns.catplot(x="Segment", y="Discount", data=df, kind="violin", palette=['red', 'pink', 'purple'])
```

The violin graph above shows the distribution of discounts across the various segments. The the graph, we receive the following insights:
1. The shapes of the three segments are identical, indicatingh that the discounts are fairly distributed across the three segments.
2. The widest parts of the violins are at 0.0 and 0.2, indicating that the majority of the sales have little to no discounts. This means that majority of the customers buy at little to no discounts, regardless of the segments.
3. There are some extensions of the violins going up to 0.8 to 0.85 (80-85% discount) in all three segments. This indicates occassional high discounts, often to clear or promote sales.

The given graph shows that most of the sales occur at low discounts (0-20%). Howwever, some occasional high sales (80-85%) exist across all segments.
